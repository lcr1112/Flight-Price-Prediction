# 2. Feature Engineering and Feature Selection

```python
df.head()
```

```python
# Categorical Variables: Airline, Source, Destination, Route, Total stops,
# Continous Variables: Duration, Price.
# Date time Variables: Date of Journet, Dep_time, Arrival_time
# Additional info is not properly defined.
# Target Variable: Price.
```

```python
df['Additional_Info'].value_counts()
```

```python
# Droping Additonal_Info and Route columns 
# As route and number of stop have same signifcance.
```

```python
df.drop(['Additional_Info','Route'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Journey_date']=pd.to_datetime(df['Date_of_Journey'],format="%d/%m/%Y").dt.day
df['Journey_month']=pd.to_datetime(df['Date_of_Journey'],format="%d/%m/%Y").dt.month
```

```python
# Converting Date of journey into Journey_date and Journey_month numeric features.
```

```python
df.head()
```

```python
df['Dep_hour']=pd.to_datetime(df['Dep_Time']).dt.hour
```

```python
df['Dep_minutes']=pd.to_datetime(df['Dep_Time']).dt.minute
```

```python
df['Arrival_hour']=pd.to_datetime(df['Arrival_Time']).dt.hour
df['Arrival_mintues']=pd.to_datetime(df['Arrival_Time']).dt.minute
```

```python
# From Dep_Time and Arrival_Time extracting hours and minutes..
```

```python
df.head()
```

```python
# Drop Arrival_Time, Dep_Time and Date_of_Journey.
df.drop(['Arrival_Time','Dep_Time','Date_of_Journey'],axis=1,inplace=True)
```

```python
df.head()
```

```python
# Conveting Airline, Source and Destination catrgorical variables into Dummey Variables
# One Hot Encoding.
```

```python
df_dummy=pd.get_dummies(df[['Airline','Source','Destination']],drop_first=True)
```

```python
df_dummy
```

```python
df.drop(['Airline','Source','Destination'],axis=1,inplace=True)
```

```python
# Assigning and converting Duration column into list
duration = list(df["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) != 2:    # Check if duration contains only hour or mins
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"   # Adds 0 minute
        else:
            duration[i] = "0h " + duration[i]           # Adds 0 hour

duration_hours = []
duration_mins = []
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))    # Extract hours from duration
    duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1]))   # Extracts only minutes from duration
```

```python
# Adding duration_hours and duration_mins list to df

df["Duration_hours"] = duration_hours
df["Duration_mins"] = duration_mins
```

```python
df.drop(["Duration"], axis = 1, inplace = True)
```

```python
df.head()
```

```python
# As no of stops increases price of flight alos increases.
#Ordinal Categorical variable 
#LabelEncoder
df.replace({"non-stop": 0, "1 stop": 1, "2 stops": 2, "3 stops": 3, "4 stops": 4}, inplace = True)
```

```python
df.head()
```

```python
plt.figure(figsize=(16,16))
sns.heatmap(df.corr(),annot=True,cmap="YlGnBu")
```

```python
# Concatenate dataframe: df and df_dummy
```

```python
df=pd.concat([df,df_dummy],axis=1)
```

```python
df.info()
```

# 3. Train Test Split

```python
y=df['Price']
```

```python
x=df.drop(['Price'],axis=1)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.3)
```

# 4. Model Building and Training

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model_RFR=RandomForestRegressor()
```

```python
model_RFR.fit(x_train,y_train)
```

```python
y_pred= model_RFR.predict(x_test)
```

```python
model_RFR.score(x_test,y_test)
```

```python
y_pred[0:12]
```

```python
y_test.head(12)
```

```python
plt.scatter(y_test, y_pred, alpha = 0.5)
plt.xlabel("y_test")
plt.ylabel("y_pred")
plt.show()
```

```python
sns.distplot(y_test,color='g',label='y_test')
sns.distplot(y_pred,color='r',label='y_pred.')
plt.legend()
```

```python
from sklearn import metrics
```

```python
MAE=metrics.mean_absolute_error(y_test,y_pred)
print(f'MAE={MAE}')
MSE=metrics.mean_squared_error(y_test,y_pred)
print(f'MSE={MSE}')
R2_score=metrics.r2_score(y_test,y_pred)
print(f'Score={R2_score}')
```
